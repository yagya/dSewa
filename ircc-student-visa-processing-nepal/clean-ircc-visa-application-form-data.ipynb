```python
#Convert dates columns with object datatype to datetime date type
def convert_str_to_date(df, column_identifier="date"):
    for idx, column_header in enumerate(df.columns.values):
        if column_identifier in column_header and not idx == 7:
            df[column_header] = pd.to_datetime(df[column_header], dayfirst=True)
```

```python
#Replace relative dates in biometrics_update_date to datetime formats
import datetime

def replace_relative_date(d, rd):
    if rd == "Same Day":
        return d
    elif rd == "The Next Day":
        return d + datetime.timedelta(days=1)
    elif type(rd) == type(datetime.datetime(2021, 9, 22)):
        return rd
    else:
        return None
```

## clean table data
1. Replace Emtpy fields with Not Specified
2. format and align college names merging multiple names of same entity to one. 


```python
#Replace any instance of empty value (NaN) in college field to a default String "Not Specified"
def replace_empty_with_NS(df, column_header, value_type="123"):
    count = 0
    ib = df[column_header]
    t = type(value_type)
    for idx, i in enumerate(ib):
        if(type(i) != t):
            df.at[idx, column_header] = "Not Specified"
            count += 1
    print(f"Processed {count} nan records and replaced with 'Not Specified' in DataFrame")

#Function to replace multiple instances of same value with single value. 
def replace_value(df, column_header, value_to_search, value_to_replace):
    count = 0
    for idx, value in enumerate(df[column_header]):
        if value_to_search.lower() in value.lower():
            if value != value_to_replace:
                df.at[idx, column_header] = value_to_replace
                count += 1
    print(f"{count} items containing {value_to_search} replaced with {value_to_replace} in DataFrame")
```
